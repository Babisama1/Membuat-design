```python
#AKHIRNYA MEMBUAT AI ASLIIIIüòÅ(WALAUPUN MASIH ACAK)
import torch.nn as nn
import torch

lapisan_saraf1= nn.Linear(in_features=1, out_features=1)
#nanti bakal keluar angka random
print(f"Lapisan pengali WEIGHT otomatis : {lapisan_saraf1.weight.item()}")
print(f"Lapisan penambah BIAS  otomatis : {lapisan_saraf1.bias.item()}")

#MELIHAT HASIL AI
input_data = torch.tensor([[5.0]])
hasil_saraf = lapisan_saraf1(input_data)

print(f"Input: {input_data.item()}")
print(f"Hasil setelah lewat saraf : {hasil_saraf.item()}")

#MELIHAT ISI OTAK AI
print("Daftar isi otak AI saat ini : ")
for name, param in lapisan_saraf1.named_parameters():
  print(f"Nama : {name}, nilai : {param.data}")
```

```python
#PELATIH AI (Stochastic Gradient Descent)
import torch.optim as optim

# 1. Reset sarafnya biar nggak pusing (bikin baru)
lapisan_saraf_fix = nn.Linear(1, 1)
#SGD
pelatih = optim.SGD(lapisan_saraf_fix.parameters(), lr=0.001)
#pengukur salah
pengukur_salah = nn.MSELoss()

#LATIHAN LOOP
input_data = torch.tensor([[1.0]])
target_benar = torch.tensor([[2.0]])

print("--MEMULAI LATIHAN--")

for putaran in range(101):
  #1. AI menebak
  tebakan = lapisan_saraf_fix(input_data)
  #2. berapa salah
  loss = pengukur_salah(tebakan, target_benar)
  #3. bersiihkan catatan lama dan hitung arah perbaikan
  pelatih.zero_grad()
  loss.backward()

  #4. pelatih mulai memperbaiki weight & bias
  pelatih.step()

  #cetak perkembangan setiap 20
  if  putaran % 20 == 0 :
      print(f"Putaran-{putaran} | Tingkat kesalahan : {loss.item():.4f}")

hasil_final = lapisan_saraf_fix(torch.tensor([[1.0]]))
print("Input : 1 | Tebakan harusnya : 2")
print(f"Tebakan AI setelah dilatih : {hasil_final.item():.2f}")
```
