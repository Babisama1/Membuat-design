```python
import torch.nn as nn
import torch.optim as optim
import torch
import torch.nn.functional as F
#MENYIAPKAN KAMUS UNTUK CHATBOT
#kamus sederhana
kata_ke_angka = {"kamu": 0, "saya": 1, "pintar": 2, "hebat": 3}
angka_ke_kata = {0: "kamu", 1: "saya", 2: "pintar", 3: "hebat" }

vocab_size = len(kata_ke_angka)
vocab_size2 = len(angka_ke_kata)

print(f"Jumlah kata : {vocab_size}")
print(f"Jumlah angka : {vocab_size2}")

#MEMBANGUN OTAK CHATBOT
class ChatbotMini(nn.Module):
  def __init__(self, vocab_size):
    super(ChatbotMini, self).__init__()
    #lapisan untuk mengubah kata jadi angka
    self.embedding = nn.Embedding(vocab_size, 10)
    #lapisan saraf
    self.saraf_pemikir = nn.Linear(10, vocab_size)

  def forward(self, x):
    x = self.embedding(x)
    y = self.saraf_pemikir(x)
    return y

model_chat = ChatbotMini(vocab_size)

#MELATIH CHATBOT(UJIAN SINGKAT)
pelatih = optim.Adam(model_chat.parameters(), lr=0.01)
pengukur_salah = nn.CrossEntropyLoss() #khusus untuk tebak kata

input_chat1 = torch.tensor([1])
target_chat1 = torch.tensor([3])

print("--CHATBOT LAGI BELAJAR--")

for i in range(50):
  prediksi = model_chat(input_chat1)
  loss = pengukur_salah(prediksi.view(1,-1), target_chat1)

  pelatih.zero_grad()
  loss.backward()
  pelatih.step()

  if i % 10 == 0:
    print(f"Putaran-{i} | Tingkat kesalahan : {loss.item():.4f}")

#TES NGOBROL BROOOðŸ˜­
"""
kata_input = "saya"
angka_input = torch.tensor([kata_ke_angka[kata_input]])

dengan_tebakan = model_chat(angka_input)
angka_hasil = torch.argmax(dengan_tebakan).item()

#tebakannya harusnya pintar
print(f"User: {kata_input}")
print(f"User: {angka_ke_kata[angka_hasil]}")"""



def tanya_ai(kata_input, suhu=1.0):
  angka_input= torch.tensor([kata_ke_angka[kata_input]])

  #AI mengeluarkan skor untuk semua kata
  skor_kata = model_chat(angka_input)

  skor_dengan_suhu = skor_kata /suhu

  peluang  = F.softmax(skor_dengan_suhu, dim=-1)

  angka_hasil = torch.multinomial(peluang, num_samples=1).item()

  return angka_ke_kata[angka_hasil]

print("Respon AI jika suhu 1.5 : ", tanya_ai("kamu", suhu=1.5))
print("Respon AI jika suhu 0.1 : ", tanya_ai("kamu", suhu=0.1))
```
