```python
#MENGAMBIL DATA SET DARI INTERNET
!pip install datasets

from datasets import load_dataset
import torch

#Kita ambil dataset puisi/teks pendek misal tiny_shakespeare

dataset_raw = load_dataset("wikitext", "wikitext-2-raw-v1", split="train")

#MEMBUAT PABRIK TOKEN
# Ambil beberapa teks dulu
text_sample = "".join(dataset_raw['text'][:1000])

# Ambil semua karakter unik
chars = sorted(list(set(text_sample)))

# Buat kamus karakter ke angka
char_to_int = {ch: i for i, ch in enumerate(chars)}
int_to_char = {i: ch for i, ch in enumerate(chars)}

print("Jumlah karakter unik:", len(chars))

def encode(s): return [char_to_int[c] for c in s]

data_angka = torch.tensor(encode(dataset_raw['text'][5]), dtype=torch.long)
print(f"Teks tadi sekarang jadi {data_angka[:20]}")

#MEMBUAT TRUK DATA LOADER
```
